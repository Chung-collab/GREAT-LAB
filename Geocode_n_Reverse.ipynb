```python
!pip install geopy
import geopy
from geopy.geocoders import Nominatim
```

```python
#!pip install ipython-autotime
#%load_ext autotime
#import pandas as pd
```

```python
#geocoder - from location address to xy coordinates
geolocator = Nominatim(timeout=10, user_agent = "myGeolocator")
location = geolocator.geocode('12 Grafton Road, Auckland')
print(location)
print((location.latitude, location.longitude))
```

```python
#reversegeocoder - from xy coordinates to location address
locator = Nominatim(timeout=10, user_agent="my-application")
coordinates = "-36.8533028, 174.7709309"
location = locator.reverse(coordinates)
print(location.latitude, location.longitude)
print(location.address)
```

```python
#show the location address structure
location.raw
```
